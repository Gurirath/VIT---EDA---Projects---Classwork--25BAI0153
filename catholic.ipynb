```python
print("\n--- Checking Missing Values ---")
print(df.isnull().sum())
```

```python
df = df[~df.hsgrad.isnull()].copy()
print("\nMissing values after cleaning max:", df.isnull().sum().max())
```

```python
print("\n--- Univariate Analysis ---")
plt.figure(figsize=(8, 5))
sns.histplot(df['read12'], kde=True)
plt.title("Distribution of 12th Grade Reading Scores")
plt.show()
```

```python
print("Skewness: %f" % df['read12'].skew())
print("Kurtosis: %f" % df['read12'].kurt())
```

```python
print("\n--- Bivariate Analysis ---")
# Scatter plot (Numeric vs Numeric)
plt.figure(figsize=(8, 5))
plt.scatter(df['lfaminc'], df['math12'], alpha=0.5)
plt.xlabel("Log Family Income (lfaminc)")
plt.ylabel("Math Score (math12)")
plt.title("Family Income vs Math Score")
plt.show()
```

```python
sns.pairplot(df, vars=['read12', 'math12', 'lfaminc'])
plt.show()
```
