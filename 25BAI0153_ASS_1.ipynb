# Exploratory Data Analysis - Sales Records Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("1000 Sales Records.csv")
```

## Basic Data Overview

```python
print("HEAD")
print(df.head(10))
print("----------------------------------------------------------")

print("TAIL")
print(df.tail(15))
print("----------------------------------------------------------")

print("INFO")
df.info()
print("----------------------------------------------------------")

print("SHAPE")
print(df.shape)
print("----------------------------------------------------------")

print("COLUMNS")
print(df.columns)
print("----------------------------------------------------------")

print("DTYPES")
print(df.dtypes)
print("----------------------------------------------------------")
```

## Descriptive Statistics

```python
print("DESCRIBE")
print(df.describe())
print("----------------------------------------------------------")

print(df.duplicated().sum())
print("----------------------------------------------------------")

print("MEAN OF FIRST 50 RECORDS")
print(df.head(50).mean(numeric_only=True))
print("----------------------------------------------------------")

print("STANDARD DEVIATION")
print(df.std(numeric_only=True))
print("----------------------------------------------------------")

print("MISSING VALUES")
print(df.isnull().sum())
print("----------------------------------------------------------")

print("UNIQUE VALUES")
print(df.nunique())
print("----------------------------------------------------------")
```

## Grouping and Sorting

```python
print("GROUP BY UNITS SOLD")
print(df.groupby("Units Sold").size())
print("----------------------------------------------------------")

print("SORTED DATA")
df_sorted = df.sort_values(
    by=["Country", "Total Revenue"],
    ascending=[True, False]
)
print(df_sorted.head(10))

print("----------------------------------------------------------")
print("USING ILOC")
print(df.iloc[0])
print("----------------------------------------------------------")
```

## Visualizations - Total Profit vs Units Sold

```python
df['Total Profit'].hist()
plt.title("Distribution of Total Profit")
plt.xlabel("Total Profit")
plt.ylabel("Frequency")
plt.show()

plt.scatter(df["Units Sold"], df["Total Profit"])
plt.xlabel("Units Sold")
plt.ylabel("Total Profit")
plt.show()
```

## Numeric Column Analysis: Units Sold

```python
print("SUMMARY OF UNITS SOLD")
print(df["Units Sold"].describe())
print("--------------------------------------------------")

print("MEAN")
print(df["Units Sold"].mean())
print("--------------------------------------------------")

print("MEDIAN")
print(df["Units Sold"].median())
print("--------------------------------------------------")

print("MODE")
print(df["Units Sold"].mode())
print("--------------------------------------------------")

print("STANDARD DEVIATION")
print(df["Units Sold"].std())
print("--------------------------------------------------")

print("MINIMUM VALUE")
print(df["Units Sold"].min())
print("--------------------------------------------------")

print("MAXIMUM VALUE")
print(df["Units Sold"].max())
print("--------------------------------------------------")
```

```python
# Histogram
plt.figure(figsize=(6,4))
df["Units Sold"].hist(bins=20)
plt.title("Distribution of Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")
plt.show()

# Box Plot
plt.figure(figsize=(5,4))
plt.boxplot(df["Units Sold"])
plt.title("Box Plot of Units Sold")
plt.ylabel("Units Sold")
plt.show()
```

## Categorical Column Analysis: Region

```python
print("REGION FREQUENCY")
print(df["Region"].value_counts())
print("--------------------------------------------------")

plt.figure(figsize=(8,5))
df["Region"].value_counts().plot(kind="bar")
plt.title("Sales Records by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## Categorical Column Analysis: Item Type

```python
print("ITEM TYPE FREQUENCY")
print(df["Item Type"].value_counts())
print("--------------------------------------------------")

plt.figure(figsize=(8,5))
df["Item Type"].value_counts().plot(kind="bar")
plt.title("Sales Records by Item Type")
plt.xlabel("Item Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
